```python
# Importing pandas
import pandas as pd
```

```python
# Reading the titanic file
df = pd.read_csv("Titanic.csv")
df.head()
```

```python
# Getting summary of df
df.info()
```

```python
# Setting the target variable (what you are trying to findout or predict)
y=df["Survived"]
```

```python
# Getting rid of no useful columns
x = df.drop(["Survived", "Name", "Ticket", "Cabin", "Embarked"], axis=1)
x
```

```python
x.info()
```

```python
# filling in the missing values in age column
x["Age"]=x["Age"].fillna(x["Age"].mode()[0])

# X[""].mode -> calculates the most frequent value of the column
# mode()[0] -> ensures we are accessing the mode value itself
# .fillna -> fills the missing values the node value
```

```python
x.info()
```

```python
# The sex column has some categorical value (words)
# Converting into numerical value as machine learning works better with numbers

x = pd.get_dummies(x,columns=["Sex"],drop_first=True)

# get_dummies -> adds a new column which will create the dummy values
# frop_first -> drops the first category , like representing male with 1 and if its not a male its a female
```

```python
df.replace({'Survived': {1: 'Yes', 0: 'No'}}, inplace=True)
y.head()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Splitting data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1, random_state=0)
```

```python
# Creating an instance of Desicion tree classifier
model = DecisionTreeClassifier()
```

```python
# Training the model
model.fit(x_train,y_train)
```

```python
# Evaluating the model
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
# Visiualizing the tree
import matplotlib.pyplot as plt
from sklearn import tree

fig = plt.figure(figsize=(100, 100))
tree.plot_tree(model,
               feature_names=x.columns.values.tolist(),
               class_names=df.Survived.unique().tolist(),
               filled=True)
fig.savefig("tree1.png")
```
